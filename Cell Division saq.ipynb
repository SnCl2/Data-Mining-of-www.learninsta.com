```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
```

```python
Url='https://www.learninsta.com/class-11-biology-important-questions-chapter-10/'
html = requests.get(Url)
soup = BeautifulSoup(html.content, 'html.parser') 
```

```python
for p in soup.find_all('p'):
    print(p.get_text())
    print("********************************")
    
```

```python
import re
def isQnA(text):
    x = re.findall('Answer:', text)
    if x :
        return True
    return False
```

```python
Qna_list=[p.get_text() for p in soup.find_all('p') if isQnA(p.get_text())]

# for p in soup.find_all('p'):
#     if isQnA(p.get_text()):
#         Qna_list.append(p.get_text())
```

```python
Q_list=[x.split("\n")[1] for x in Qna_list]
print(*Q_list)
```

```python
A_list=[x.split("\n")[-1] for x in Qna_list]
print(*A_list)
```

```python
df = pd.DataFrame(list(zip(Q_list, A_list)),
               columns =['Q', 'A'])
df
```
